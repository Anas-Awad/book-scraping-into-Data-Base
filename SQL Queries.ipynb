```python
import sqlite3
import pandas as pd
connection = sqlite3.connect('books.db')
```

# 01) Write a sql query to get books that has rate >=3 and has 'Mr' in its name.

```python
query = """
select * 
from books
where book_rate >= 3 and book_name like '%Mr%'
"""
data = connection.execute(query).fetchall()
print(data)
```

# 14) Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.

```python
query = """
select 
    book_name, book_rate, book_price, category_name
from books
join categories
    using(category_id)
"""
data = connection.execute(query).fetchall()
for i in data:
    print(i)
```

# 15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.

```python
query = """
select book_rate , count(book_name) number_of_books
from books
where book_price between 20 and 30 
group by book_rate
"""
data = connection.execute(query).fetchall()
for i in data:
    print(i)
```

# 16) Write a sql query to calculate how many books each category has having count > 10.

```python
query = """
select 
    count(book_name) , category_name
from books
join categories
    using(category_id)
group by category_name
having count(book_name) > 10
"""
data = connection.execute(query).fetchall()
for i in data:
    print(i)
```

# 17) Write a sql query to get all books with category_name='Music' using subquery.

```python
query = """
select book_name , 'Music' as category
from books 
where category_id = (select category_id from categories where category_name  = 'Music')
"""
data = connection.execute(query).fetchall()
for i in data:
    print(i)
```
